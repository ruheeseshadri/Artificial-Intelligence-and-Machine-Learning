Question 2

Given a random sample of 10 manufactured items is selected. 

Therefore, n = 10. The failure rate given is 5%. We can say it follows a binomial distribution with failure rate = 0.05 and n = 10.

```python
# Importing library for probability distribution functions
import scipy.stats as stats
```

```python
# Given
p = 0.05
n = 10
```

```python
# Q2) A) Probability that no items are defective
Required_Probability = stats.binom.pmf(0, n, p)
Required_Probability
```

Probability that no items are defective is 0.5987369392383787.

```python
# Q2) B) Probability that exactly one of the items is defective
Required_Probability = stats.binom.pmf(1, n, p)
Required_Probability
```

Probability that exactly one of the items is defective is 0.3151247048623047.

```python
# Q2) C) Probability that two or fewer of the items are defective
Required_Probability = stats.binom.pmf(0, n, p) + stats.binom.pmf(1, n, p) + stats.binom.pmf(2, n, p) 
Required_Probability
```

Probability that two or fewer of the items are defective is 0.988496442620703.

```python
# Q2) D) Probability that three or more of the items are defective = 1 - Probability that two or fewer of the items are defective
Required_Probability = 1 - (stats.binom.pmf(0, n, p) + stats.binom.pmf(1, n, p) + stats.binom.pmf(2, n, p))
Required_Probability
```

Probability that three or more of the items are defective is 0.01150355737929698.

Question 3

Given a salesman sells 3 cars per week on a average. Therefore, it follows a poisson distribution with lambda = 3.

```python
# Importing poisson distribution
from scipy.stats import poisson
```

```python
# Given
lambda_val = 3
```

```python
# Q3) A) Probability that in a given week he will sell some cars = 1 - Probability that in a given week he will sell no cars
Required_Probability = 1 - poisson.pmf(0, lambda_val)
Required_Probability
```

Probability that in a given week he will sell some cars is 0.950212931632136.

```python
# Q3) B) Probability that in a given week he will sell 2 or more but less than 5 cars = Probability that in a given week he will sell 2 or 3 or 4 cars
Required_Probability = poisson.pmf(2, lambda_val) + poisson.pmf(3, lambda_val) + poisson.pmf(4, lambda_val)
Required_Probability
```

Probability that in a given week he will sell 2 or more but less than 5 cars is 0.6161149710523164.

```python
# Q3) C) Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week
# Importing the required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
x = np.arange(0, 21)
cumulative_prob = poisson.cdf(x, lambda_val)
plt.plot(x, cumulative_prob)
plt.xlabel('Number of Cars sold per-week')
plt.ylabel('Cumulative Probability of Cars sold per-week')
plt.title('Poisson Distribution Function')
plt.show()
```

```python
# Checking for null values
data.isnull().sum()
```

'Product' and 'Funding' have 6 and 214 null values respectively.

```python
# Dropping the null values
new_data = data.dropna(subset = ['Product', 'Funding'])
new_data
```

```python
# Checking if there are any null values
new_data.isnull().sum()
```

```python
# Converting 'Funding' to a numerical value
new_data.loc[:,'Funds_in_million'] = new_data['Funding'].apply(lambda x: float(x[1:-1])/1000 if x[-1] == 'K' else (float(x[1:-1])*1000 if x[-1] == 'B' else float(x[1:-1])))
```

```python
new_data
```

```python
# Importing libraries for plotting
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Plotting a box plot for 'Funds_in_million'
sns.boxplot(data = new_data, x = 'Funds_in_million')
plt.title('Box Plot for Funds')
plt.show()
```

```python
# Checking number of outliers greater than the upper fence
Q1 = new_data['Funds_in_million'].quantile(0.25)
Q3 = new_data['Funds_in_million'].quantile(0.75)
IQR = Q3 - Q1
lower_whisker = Q1 - 1.5*IQR
upper_whisker = Q3 + 1.5*IQR
((new_data['Funds_in_million'] > upper_whisker)).sum()
```

There are 60 outliers greater than the upper fence.

```python
# Frequency of 'OperatingState' features classes
new_data['OperatingState'].value_counts()
```

```python
# Importing the required functions for statistical analysis
from scipy import stats
from scipy.stats import ttest_ind
from statsmodels.stats.proportion import proportions_ztest
```

```python
# Plotting a histogram to see if there is any visual difference between funds raised by companies that are still operating vs companies that closed down
fig, axes = plt.subplots(1, 2, figsize=(10,5))
sns.histplot(new_data.loc[new_data.OperatingState == 'Operating', 'Funds_in_million'], ax = axes[0], kde = True)
sns.histplot(new_data.loc[new_data.OperatingState == 'Closed', 'Funds_in_million'], ax = axes[1], kde = True)
axes[0].set_title('Funds raised by companies that are still operating')
axes[1].set_title('Funds raised by companies that closed down')
plt.show()
```

Both distributions are right skewed. There is not much visual difference.

Null Hypothesis (Ho) : There is no significant difference between the 2 means.

Alternate Hypothesis (Ha) : There is significant difference between the 2 means.

```python
t_statistic, p_value  = ttest_ind(new_data[new_data['OperatingState'] == 'Operating']['Funds_in_million'], new_data[new_data['OperatingState'] == 'Closed']['Funds_in_million'])

# Let level of significance be 5%
alpha = 0.05

print('tstat',t_statistic)
print('p value',p_value)
```

```python
if p_value <= alpha:
    print('We have enough evidence to reject the null hypothesis in favour of alternate hypothesis.')
    print('We conclude that the mean difference between the 2 samples is same.')
else:
    print('We do not have enough evidence to reject the null hypothesis in favour of alternate hypothesis.')
    print('We conclude that mean difference between the 2 samples is not same.')
```

```python
# Creating a copy of original data
copy_data = new_data.copy()
copy_data
```

```python
# Frequency of 'Result' variables
copy_data['Result'].value_counts()
```

```python
# Percentage of winners that are still operating and percentage of contestants that are still operating
winners = copy_data['Result'].value_counts()['Winner']
contestants = copy_data['Result'].value_counts()['Contestant']
contestants_operating = copy_data['OperatingState'][copy_data['Result'] == 'Contestant'].value_counts().loc['Operating']
winners_operating = copy_data['OperatingState'][copy_data['Result'] == 'Winner'].value_counts().loc['Operating']
(winners_operating/winners)*100, (contestants_operating/contestants)*100
```

Percentage of winners that are still operating : 72% 

Percentage of contestants that are still operating : 68.59%

Hypothesis : There is difference between the proportion of companies that are operating in the winner and contestant category.

Null Hypothesis (Ho) : Proportion of companies that are operating is same in both categories.

Alternate Hypothesis (Ha) : Proportion of companies that are operating is different in both categories.

```python
t_statistic, p_value = proportions_ztest([contestants_operating, winners_operating], [contestants, winners])

# Let level of significance be 5%
alpha = 0.05

print('tstat', t_statistic)
print('p-value', p_value)
```

```python
if p_value <= alpha:
    print('We have enough evidence to reject the null hypothesis in favour of alternate hypothesis.')
    print('We conclude that the proportion of companies that are operating is same in both categories.')
else:
    print('We do not have enough evidence to reject the null hypothesis in favour of alternate hypothesis.')
    print('We conclude that the proportion of companies that are operating is different in both categories.')
```
